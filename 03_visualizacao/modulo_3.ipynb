## Tópico 3.2: Customização Simples (Sem Complicação!)

Você não precisa de códigos longos para deixar o gráfico com cara de relatório executivo. Podemos usar os recursos nativos do Seaborn para agrupar informações rapidamente.

Vamos voltar ao nosso gráfico de vendas, mas agora queremos ver as vendas divididas por **Ano** e por **Região** no mesmo gráfico.

## Bônus: O Mundo "Clássico" do Python (Matplotlib)

Você deve estar se perguntando: *"Se o Seaborn é tão fácil, por que as pessoas usam outras coisas?"*

A verdade é que o Seaborn é construído **em cima** de uma biblioteca muito mais antiga e famosa chamada **Matplotlib**. O Matplotlib é extremamente poderoso, mas ele é "manual".

Para provar como o Seaborn facilita a sua vida nos negócios, veja a diferença para criar um simples gráfico de barras com a média de vendas por região:

```python
# O JEITO SEABORN (Automático)
# Ele mesmo já junta as regiões iguais e calcula a média das vendas sozinho!
sns.barplot(data=df_vendas, x='Regiao', y='Faturamento', color='blue')
```

```python
# O JEITO MATPLOTLIB (Manual)
import matplotlib.pyplot as plt

# 1. Primeiro precisamos agrupar os dados e calcular a média manualmente (tipo Tabela Dinâmica)
media_vendas = df_vendas.groupby('Regiao')['Faturamento'].mean()

# 2. Depois temos que passar os eixos separadamente para desenhar o gráfico
plt.bar(media_vendas.index, media_vendas.values, color='gray')
plt.show()
```

**Resumo para Negócios:**
Use o **Matplotlib** quando precisar desenhar do zero ou fazer ajustes muito finos.
Use o **Seaborn** para transformar suas tabelas em *insights* visuais estatísticos em segundos!

## 📊 Exemplo de Negócio — Módulo 3

**Cenário:** O time de marketing quer uma visualização rápida para apresentar na reunião de resultados. 
Você precisa mostrar a **relação entre investimento em marketing e vendas geradas**, 
com um gráfico de dispersão profissional.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='whitegrid')

df_mkt = pd.read_csv('https://raw.githubusercontent.com/gustavocxavier/analise-de-dados-para-negocios/refs/heads/main/Dados/marketing.csv')

# Gráfico de dispersão com linha de tendência
sns.regplot(data=df_mkt, x='Investimento', y='Vendas',
            scatter_kws={'alpha':0.5}, line_kws={'color':'red'})

plt.title('Investimento em Marketing vs. Vendas Geradas', fontsize=14)
plt.xlabel('Investimento (R$)')
plt.ylabel('Vendas (R$)')
plt.tight_layout()
plt.show()
```

**Interpretação:** A linha vermelha é a **linha de tendência (regressão linear)**. 
Se os pontos acompanham bem a linha, o investimento em marketing tem um impacto consistente nas vendas. 
Pontos muito distantes da linha são campanhas com resultado anormal (muito acima ou abaixo do esperado).
