## 2.1 Importando Dados

A primeira etapa de qualquer projeto de negócios é carregar as informações. Vamos usar a função `read_csv()` do Pandas para ler uma base de dados imobiliários que já vem de exemplo no Google Colab.

````python
# 1. Importando a biblioteca Pandas (apelidada carinhosamente de 'pd')
import pandas as pd

# 2. Lendo o arquivo CSV e guardando em uma variável chamada 'df' (Data Frame)
df = pd.read_csv('sample_data/california_housing_train.csv')

# 3. Mostrando as 5 primeiras linhas da nossa tabela (como se fosse o topo da planilha)
df.head()
````

## 2.2 Inspeção e Limpeza

Dados reais são bagunçados! Eles vêm com informações faltando (valores nulos ou `NaN`) e linhas duplicadas. Se não limparmos isso, nossas análises e decisões de negócios estarão erradas.

Vamos inspecionar a saúde da nossa base.


````python
# Quantas linhas e colunas temos?
print("Tamanho da base:", df.shape)

# Existe algum valor vazio (nulo) na nossa tabela?
print("\nValores vazios por coluna:")
print(df.isnull().sum())
````

Como sabemos que dados sujos atrapalham decisões, vamos usar alguns comandos para tratar os valores nulos e remover duplicatas.

````python
# LIMPEZA DE DADOS

# Removendo linhas que tenham valores vazios (Drop NA)
df_limpo = df.dropna()

# Removendo linhas idênticas (Duplicatas)
df_limpo = df_limpo.drop_duplicates()

print("Tamanho original:", df.shape)
print("Limpeza concluída! Nova quantidade de linhas:", df_limpo.shape[0])
````

## Bônus: O Mundo "Clássico" do Python (Matplotlib)

Você deve estar se perguntando: *"Se o Seaborn é tão fácil, por que as pessoas usam outras coisas?"*

A verdade é que o Seaborn é construído **em cima** de uma biblioteca muito mais antiga e famosa chamada **Matplotlib**. O Matplotlib é extremamente poderoso, mas ele é "manual".

Para provar como o Seaborn facilita a sua vida nos negócios, veja a diferença para criar um simples gráfico de barras com a média de vendas por região:

````python
# O JEITO SEABORN (Automático)
# Ele mesmo já junta as regiões iguais e calcula a média das vendas sozinho!
sns.barplot(data=df_vendas, x='Regiao', y='Faturamento', color='blue')
````

````python
# O JEITO MATPLOTLIB (Manual)
import matplotlib.pyplot as plt

# 1. Primeiro precisamos agrupar os dados e calcular a média manualmente (tipo Tabela Dinâmica)
media_vendas = df_vendas.groupby('Regiao')['Faturamento'].mean()

# 2. Depois temos que passar os eixos separadamente para desenhar o gráfico
plt.bar(media_vendas.index, media_vendas.values, color='gray')
plt.show()
````

**Resumo para Negócios:**
Use o **Matplotlib** quando precisar desenhar do zero ou fazer ajustes muito finos.
Use o **Seaborn** para transformar suas tabelas em *insights* visuais estatísticos em segundos!

# Módulo 4: Testes de Hipóteses com Pingouin

Em negócios, nós não gostamos de adivinhar. Se a nova página do site (Versão B) vendeu mais que a antiga (Versão A), foi porque ela é realmente melhor ou foi apenas *sorte* daquela semana?

A estatística responde isso. Vamos usar a biblioteca **Pingouin** para fazer testes rigorosos de forma extremamente simples.

> **Nota:** O Pingouin não vem instalado por padrão no Google Colab. Vamos instalá-lo com o comando abaixo (o `!` indica que é um comando do sistema operacional, não Python).

````python
# Instalando o Pingouin (execute esta célula apenas uma vez por sessão do Colab)
!pip install pingouin
import pingouin as pg
````

## Tópico 4.1: Teste A/B (O Famoso Teste T)

**Cenário:** Lançamos duas versões de um anúncio (A e B). Queremos saber se a diferença de cliques entre eles é real.

**Regra de Ouro da Estatística para Negócios:**
Olhe para a coluna **`p-val`** (Valor-P) no resultado.
* Se for **MENOR que 0.05**: A diferença é real e comprovada!
* Se for **MAIOR que 0.05**: A diferença não é confiável (pode ter sido sorte).

````python
# Lendo os dados do nosso Teste A/B
df_ab = pd.read_csv('teste_ab_vendas.csv')

# Separando quem viu o Anúncio A e quem viu o Anúncio B
vendas_A = df_ab[df_ab['Anuncio'] == 'A']['Vendas']
vendas_B = df_ab[df_ab['Anuncio'] == 'B']['Vendas']

# Executando o Teste T com Pingouin
pg.ttest(x=vendas_A, y=vendas_B)
````

## Tópico 4.2: Comparando Várias Opções (ANOVA)

O Teste T só compara duas coisas. E se quisermos testar **3 designs diferentes de embalagem** nas prateleiras e ver qual vende mais?

Para comparar 3 ou mais grupos, usamos a **ANOVA** (Análise de Variância). O Pingouin facilita muito porque podemos passar a tabela inteira para ele.

````python
# Lendo dados de vendas de 3 designs diferentes (Vermelho, Azul, Verde)
df_embalagens = pd.read_csv('embalagens.csv')

# Executando a ANOVA:
# data = nossa tabela
# dv = variável dependente (o que queremos medir? 'Vendas')
# between = qual coluna divide os grupos? ('Design_Embalagem')

pg.anova(data=df_embalagens, dv='Vendas', between='Design_Embalagem')
````

## Tópico 4.3: Correlação (Causa e Efeito?)

Queremos provar matematicamente se dar descontos maiores realmente aumenta o volume de compras.

A Correlação vai de -1 a 1. Quanto mais perto de 1, mais forte é a relação positiva (um sobe, o outro sobe).

````python
# Lendo os dados de promoções
df_promocoes = pd.read_csv('descontos.csv')

# Testando a correlação entre a porcentagem de desconto e as unidades vendidas
pg.corr(x=df_promocoes['Desconto_Oferecido'], y=df_promocoes['Unidades_Vendidas'])
````

## Bônus: O Mundo "Clássico" do Python (SciPy)

Por que escolhemos o Pingouin para este curso?

No mercado de trabalho ou lendo tutoriais no Google, você fatalmente esbarrará no pacote **`scipy.stats`**. Ele é a ferramenta mais antiga e clássica para estatística em Python.

Vamos rodar o mesmo Teste A/B do início usando o `scipy` apenas para você ver a diferença. Note como a resposta dele é feita para matemáticos, enquanto o Pingouin gera uma tabela bonita e fácil de ler (DataFrame).

````python
# Importando a biblioteca clássica
from scipy import stats

# Rodando o mesmo Teste T (Teste A/B) do Tópico 4.1
resultado_scipy = stats.ttest_ind(vendas_A, vendas_B)

# Veja como a apresentação do resultado é mais "crua"
print(resultado_scipy)
````
