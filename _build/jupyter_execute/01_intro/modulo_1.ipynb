### 1.3.1. Séries Temporais (Gráfico de Linha)

Vamos analisar um conjunto de dados real sobre a quantidade de passageiros em voos ao longo dos anos. Veja como é simples criar um gráfico de série temporal com apenas uma linha de código principal.

````python
# Carregando os dados de voos que já vem embutidos no Seaborn
voos = sns.load_dataset("flights")

# Criando o gráfico de linha (Série Temporal) com xlabel e ylabel
sns.lineplot(data=voos, x="year", y="passengers")
plt.xlabel("Ano")
plt.ylabel("Passageiros")
````

### 1.3.2. Analisando Categorias (Boxplot)

Agora, vamos imaginar que você gerencia um restaurante. Você quer saber: **Em qual dia da semana os clientes gastam mais dinheiro?**

O *Boxplot* é o gráfico perfeito para relacionar uma categoria (Dia da Semana) com um número contínuo (Valor da Conta). Ele mostra a mediana, os quartis e os valores atípicos (outliers) de forma compacta.

````python
# Carregando dados de gorjetas e contas de um restaurante
gorjetas = sns.load_dataset("tips")

# Criando o Boxplot
sns.boxplot(data=gorjetas, x="day", y="total_bill", hue="day", palette="Set2", legend=False)
````

## 📊 Exemplo de Negócio — Módulo 1

**Cenário:** Você é analista em uma distribuidora. O gestor quer saber rapidamente o **lucro líquido** de uma venda específica e pede que você crie um mini-script em Python para calcular e exibir o resultado.

Execute a célula abaixo e observe como o Python funciona como uma calculadora inteligente.


````python
# Exemplo de Negócio – Módulo 1
# Dados de uma venda da distribuidora
preco_venda   = 4500.00  # Receita da venda
custo_produto = 2800.00  # Custo do produto
frete         = 150.00   # Custo de frete
impostos      = 360.00   # Impostos sobre a venda

lucro_liquido = preco_venda - custo_produto - frete - impostos

print(f'Receita:          R$ {preco_venda:,.2f}')
print(f'Custos totais:    R$ {(custo_produto+frete+impostos):,.2f}')
print(f'Lucro líquido:    R$ {lucro_liquido:,.2f}')
print(f'Margem de lucro:  {lucro_liquido/preco_venda*100:.1f}%')
````

**Interpretação:** A célula acima mostra a margem de lucro daquela venda. 
Se a margem cair abaixo de 10%, pode ser necessário renegociar o custo do produto ou o frete.
