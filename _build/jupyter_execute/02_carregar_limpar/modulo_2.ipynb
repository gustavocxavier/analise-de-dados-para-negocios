## 2.1 Importando Dados

A primeira etapa de qualquer projeto de negócios é carregar as informações. Vamos usar a função `read_csv()` do Pandas para ler uma base de dados imobiliários que já vem de exemplo no Google Colab.

> **Dica:** Em projetos reais, você substituiria `'sample_data/california_housing_train.csv'` pelo caminho do seu próprio arquivo, por exemplo `'vendas_2024.csv'`.


```python
# 1. Importando a biblioteca Pandas (apelidada carinhosamente de 'pd')
import pandas as pd

# 2. Lendo o arquivo CSV e guardando em uma variável chamada 'df' (Data Frame)
# df = pd.read_csv('sample_data/california_housing_train.csv')

# Lendo diretamente de um site
df = pd.read_csv('https://raw.githubusercontent.com/gustavocxavier/analise-de-dados-para-negocios/refs/heads/main/Dados/vendas.csv')

# 3. Mostrando as 5 primeiras linhas da nossa tabela (como se fosse o topo da planilha)
df.head()
```

## 2.2 Inspeção e Limpeza

Dados reais são bagunçados! Eles vêm com informações faltando (valores nulos ou `NaN`) e linhas duplicadas. Se não limparmos isso, nossas análises e decisões de negócios estarão erradas.

Vamos inspecionar a saúde da nossa base.


```python
# Quantas linhas e colunas temos?
print("Tamanho da base:", df.shape)

# Existe algum valor vazio (nulo) na nossa tabela?
print("\nValores vazios por coluna:")
print(df.isnull().sum())
```

Como sabemos que dados sujos atrapalham decisões, vamos usar alguns comandos para tratar os valores nulos e remover duplicatas.

```python
# LIMPEZA DE DADOS

# Removendo linhas que tenham valores vazios (Drop NA)
df_limpo = df.dropna()

# Removendo linhas idênticas (Duplicatas)
df_limpo = df_limpo.drop_duplicates()

print("Tamanho original:", df.shape)
print("Limpeza concluída! Nova quantidade de linhas:", df_limpo.shape[0])
```

## 2.3 Estatística Descritiva

Com os dados limpos, podemos começar a extrair inteligência. Qual é o preço médio dos imóveis? Qual é o valor máximo?

O Pandas nos dá um resumo executivo com apenas um comando: `.describe()`. Depois, vamos preparar o terreno para estatísticas mais avançadas com o pacote **Pingouin**.

```python
# O resumo executivo dos dados numéricos (Média, Desvio Padrão, Min, Max)
# O comando describe() gera a estatística de todas colunas numéricas de uma vez!
df_limpo.describe()
```

## 📊 Exemplo de Negócio — Módulo 2

**Cenário:** O departamento de RH precisa de um relatório rápido sobre a folha de pagamento. 
Vamos carregar o arquivo `rh_salarios.csv`, inspecionar a saúde dos dados e extrair estatísticas descritivas para uma reunião executiva.


```python
import pandas as pd

# Carregando dados de salários do RH
df_rh2 = pd.read_csv('https://raw.githubusercontent.com/gustavocxavier/analise-de-dados-para-negocios/refs/heads/main/Dados/rh_salarios.csv')

# Inspeção rápida
print('Linhas e colunas:', df_rh2.shape)
print('Valores nulos:\n', df_rh2.isnull().sum())

# Limpeza
df_rh2 = df_rh2.dropna().drop_duplicates()

# Estatísticas executivas
print('\n--- Relatório de Salários ---')
print(f"Salário médio:   R$ {df_rh2['Salario'].mean():,.2f}")
print(f"Salário mediano: R$ {df_rh2['Salario'].median():,.2f}")
print(f"Menor salário:   R$ {df_rh2['Salario'].min():,.2f}")
print(f"Maior salário:   R$ {df_rh2['Salario'].max():,.2f}")
```

**Interpretação:** Quando a média é muito maior que a mediana, existe uma concentração de salários 
altos 'puxando' a média. Nesse caso, a **mediana** é a medida mais honesta para comunicar ao RH 
o salário 'típico' da empresa.
